## 1. Data Exploration with Pandas
Data exploration involves examining datasets to understand their structure, contents, and key statistics. This step is critical in preparing data for further analysis and visualization.

### Common Functions:
- `.head()` - Display the first few rows of the dataset.
- `.describe()` - Generate summary statistics for numerical columns.
- `.info()` - Display information about the datasetâ€™s structure.
- `.value_counts()` - Count the occurrences of unique values in a column.

Let's start by loading a dataset and exploring its structure.


```python

# Importing pandas and loading a sample dataset
import pandas as pd

# Sample dataset
data = {
    'Date': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04'],
    'Visitors': [200, 150, 300, 100],
    'Location': ['Park A', 'Museum B', 'Park A', 'Beach C'],
    'Revenue': [1000, 750, 1500, 500]
}

df = pd.DataFrame(data)

# Exploring the dataset
df.head()
```

```python

# Display basic statistics
df.describe()

# Display dataset structure
df.info()
```


## 2. Data Visualization with Matplotlib and Seaborn
Once the data is explored and understood, the next step is visualization. Visualization helps reveal patterns, trends, and insights in the data.

### Key Visualization Techniques:
- **Bar Charts**: Compare different categories.
- **Line Plots**: Show trends over time.
- **Histograms**: Display the distribution of data.

We will use two Python libraries for visualization:
- `matplotlib`: A powerful plotting library for basic visualizations.
- `seaborn`: Built on top of `matplotlib`, it provides additional functionality and better aesthetics.

### Example 1: Bar Chart to Visualize Visitors by Location


```python

# Importing matplotlib for visualization
import matplotlib.pyplot as plt

# Bar chart: Visitors by Location
df.groupby('Location')['Visitors'].sum().plot(kind='bar')

# Add labels and title
plt.title('Total Visitors by Location')
plt.xlabel('Location')
plt.ylabel('Visitors')
plt.show()
```


### Example 2: Line Plot to Visualize Revenue Over Time
A line plot helps show trends in the data over a period. In this case, we will plot the revenue generated over the course of four days.


```python

# Line plot: Revenue over time
df.plot(x='Date', y='Revenue', color='green', marker='o', linestyle='dashed')
# https://matplotlib.org/stable/api/markers_api.html
# Add labels and title
plt.title('Revenue Over Time')
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.show()
```


### Example 3: Histogram to Visualize Distribution of Visitors
A histogram helps us understand the distribution of numerical data. We will plot the distribution of the number of visitors across different locations.


```python

# Histogram: Distribution of Visitors
df['Visitors'].plot(kind='hist', bins=5)

# Add labels and title
plt.title('Distribution of Visitors')
plt.xlabel('Number of Visitors')
plt.ylabel('Frequency')
plt.show()
```
